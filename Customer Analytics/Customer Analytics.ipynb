### Standardization

```python
scaler = StandardScaler()
segmentation_std = scaler.fit_transform(df_segmentation)
```

### Hierarchical Clustering

```python
# Hierarchical Clustering for Segmentation and Customer Analytics
# Dendogram - tree-like, hierarchical representation of points
# Linkage - specific cluster methd we employ
```

```python
hier_clust = linkage(segmentation_std, method = 'ward')
# Returns a linkage matrix
# Represent linkage matrix like a Dendrogram
```

```python
# Removing the color threshold
plt.figure(figsize=(12,9))
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hier_clust,
           truncate_mode='level',
           p=5,
          show_leaf_counts=False,
          no_labels=True)
plt.show()
```
